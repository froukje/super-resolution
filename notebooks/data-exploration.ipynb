# Setup

```python
#from PIL import Image
import glob 
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

# Load Images and plot some Examples

```python
path_lr = '../data/LR'
path_hr = '../data/HR'
```

```python
#path_lr = '../data/Image_Super_Resolution/dataset/Raw Data/low_res'
#path_hr = '../data/Image_Super_Resolution/dataset/Raw Data/high_res'
```

```python
imgs_lr = glob.glob(f"{path_lr}/*.png")
imgs_hr = glob.glob(f"{path_hr}/*.png")

imgs_lr.sort()
imgs_hr.sort()
```

```python
print(f'nr of low res images {len(imgs_lr)}')
print(f'nr of high res images {len(imgs_hr)}')
```

```python
print(imgs_lr[:5])
```

```python
print(imgs_hr[:5])
```

```python
idxs_all = np.arange(len(imgs_lr))
idxs_sel = random.choices(idxs_all, k=10)

fig, axes = plt.subplots(2, 4, figsize=(20,10))

for i in range(4):
    lr = mpimg.imread(imgs_lr[idxs_sel[i]])
    hr = mpimg.imread(imgs_hr[idxs_sel[i]])
 
    print(f'low resolution shape: {lr.shape}')
    print(f'high resolution shape: {hr.shape}')
    axes[0,i].imshow(lr)
    axes[1,i].imshow(hr)
    axes[0,i].set_xticks([])
    axes[0,i].set_yticks([])
    axes[1,i].set_xticks([])
    axes[1,i].set_yticks([]);
```

```python
print(f'High resolution images have {384/96} higher resolution than lower resolution images')
```
